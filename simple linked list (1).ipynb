# Simple implementation

```c++
class Node{
    public:
    int data;
    Node* next;
    Node(int x){
        data=x;
        next=NULL;
    }
}
```

```c++
class linkedList{
    protected:
    Node *head;
    public:
    linkedList(){
        head=NULL;
    };
    
    void push(int x);
    void append(int x);
    void printlist();
}
```

### insertertion at beginging

```c++
void linkedList::push(int x){
Node *temp=new Node(x);
temp->next=head;
head=temp;
}
```

### insertertion at end

```c++
void linkedList::append(int x){
    Node *last=head;
    Node *temp=new Node(x);
    while(last->next!=NULL){
        last=last->next;
    }
    last->next=temp;
}
```

### displaying the list

```c++
void linkedList::printlist(){
    Node* curr=head;
    while(curr!=NULL){
        cout<<(curr->data)<<" ";
        curr=curr->next;
    }
}
```

```c++
linkedList l;
l.push(2);
l.push(3);
l.push(4);
```

```c++
l.printlist();
```

```c++
l.append(3);
l.printlist();
```

# insertertion

```c++
Node *insertBegin(Node *head,int x){
    Node *temp=new Node(x);
    temp->next=head;
    return temp;
}
```

```c++
Node *push(Node *head,int x){
    Node *temp=head;
    head->next=temp;
    head->data=x
    return;
}
```
