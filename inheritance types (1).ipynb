```c++
#include <iostream>
#include <string>
using namespace std;
```

##### single inheritence

```c++
class laptop{
    public:
    int price;
    int ram;
}
```

```c++
class device : public laptop{
    public:
    void display() {
        cout <<"\nThis is a device";
    }
}
```

```c++
device obj1;
obj1.price=40000;
obj1.ram=5;
cout <<"device cost = "<<obj1.price;
cout <<"\ndevice ram = "<<obj1.ram;
obj1.display()
```

##### multiple inheritance

```c++
class laptop{
    protected:
    int cost=40000;
    int storage = 512;
    int ram=8;
    string type="gaming laptop";
    void 
    {
        cout << "laptop details : " << endl; 
        cout << "ram : " << ram << endl; 
        cout << "storage : " << storage << endl;
        cout << "laptop  type : " << type << endl;
        cout << "cost : " << cost << endl; 
    }
}
```

```c++
class mobile{
    protected:
    int x;
    int cost=10000;
    int storage = 64;
    int ram=4;
    string display = "amoled";
    void mobile_details(){
        cout << "mobile details : " << endl; 
        cout << "ram : " << ram << endl; 
        cout << "storage : " << storage << endl;
        cout << "display type : " << display << endl;
        cout << "cost : " << cost << endl; 
    }
    
}
```

```c++
class device : public laptop, public mobile{
    public:
    void display(){
        mobile_details();
        cout  << endl;
        laptop_details();
    }
    
}
```

```c++
device dvc;
dvc.display()
```

##### mutlilevel inheritance

```c++
class house_interior{
    protected:
    int cupboards = 5;
    int beds = 6;
}
```

```c++
class house_exterior : public house_interior{
    protected:
    string type = "2BHK";
    int cost = 5000000;
}
```

```c++
class house_details : public house_exterior{
    public:
    void display(){
        cout << "No. of cupboards : " << cupboards << endl;
        cout << "No. of bed : " << cupboards << endl; 
        cout << "Types : " << cupboards << endl; 
        cout << "Cost : " << cupboards << endl; 
    }
}
```

```c++
house_details home;
home.display();
```
